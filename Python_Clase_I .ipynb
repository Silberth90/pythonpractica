##üí°Bucle **While**

####üí°El bucle while eval√∫a una condici√≥n y luego ejecuta un bloque de c√≥digo si la condici√≥n es verdadera.

####üí°El bloque de c√≥digo se ejecuta repetidamente hasta que la condici√≥n llega ser o es falsa.

```python
 while condicion :
  #ejecuta codigo
```

```python
entrada = int(input('Ingresa un n√∫mero bro: '))
maximo = 100
if entrada != 0:
    while entrada < maximo:
        print(f'entrada{entrada}')
        if entrada <= maximo/2:
            entrada = entrada * 2
            print(f'El n√∫mero multiplicado por 2 es {entrada}')
        else:
            break
else:
    print('Revisa el dato bro')
```

####‚ö°Y con un contador?


```python
contador=0
numero=2
iteracionesMax=2

while contador<=iteracionesMax:
  numero=numero*2
  print(f'Numero:{numero}')
  print(f'Contador:{contador}')
  contador+=1
```
